```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so

import random as rd

#Create list 
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health'] 

#number of enries 
n = 500


# Generate random data
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [rd.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Using value_counts on Category column
category_counts = df['Category'].value_counts()

# Print DataFrame Information 
print(df.info())

# Print DataFrame Description 
print(df.describe())

# Print the first few rows of the DataFrame
print(df.head(10))


print(category_counts)


null_values = df.isnull()
```

```python
# Removing Null Values from DataFrame 
df_nunNull = df.dropna()

# Removing duplicate data using drop_duplicates()
df_cleaned = df_nunNull.drop_duplicates()

# Convert the 'Date' column to datetime format using to_datetime
df_cleaned['Date'] = pd.to_datetime(df['Date'])


df_cleaned['Likes'] = df_cleaned['Likes'].astype(int)

print(df_cleaned.info())
#print(df.info())
```

```python
sns.histplot(df_cleaned['Likes'], bins = 8  )

# Set plot labels and title for the histogram
plt.xlabel('Category')
plt.ylabel('Likes')

plt.show()
```

```python
# Create a boxplot with 'Category' on the x-axis and 'Likes' on the y-axis
sns.boxplot(x='Category', y='Likes', data=df_cleaned)

# Set plot labels and title for the boxplot
plt.xlabel('Category')
plt.ylabel('Likes')
plt.title('Boxplot of Likes by Category')

# Show the boxplot
plt.show()
```
